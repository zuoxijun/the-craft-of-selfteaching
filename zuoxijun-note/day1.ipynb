```python
import matplotlib.pyplot as plt
import numpy as np
```

## 平均寿命预期

```python
data = np.genfromtxt(fname="../life-expectancy-china-1960-2016.txt", delimiter=",", names=['x', 'y'])
```

```python
ex1960 = data[0][1]
ex2016 = data[-1][1]
increase_rate = (ex2016 - ex1960) / ex1960
```

```python
note = "from {:.2f} in 1960 to {:.2f} in 2016, increased {:.2%}".format(ex1960, ex2016, increase_rate)
```

```python
plt.figure(figsize=(10, 5))    # 设置画布的长和宽
plt.plot(data['x'], data['y']) # 指定用于绘制横坐标和纵坐标的数据
plt.title("CHINA\n" + note)    # 指定图像title
plt.ylabel("Lift Expectancy from Birth") # 指定轴的title
plt.tick_params(axis='x', rotation=70)  # 指定对哪一个轴进行操作  这里指定将x的刻度逆时针旋转70度
plt.show()
```

## 人类发展指数

```python
lifeData = np.genfromtxt("../life-expectancy-china-1960-2016.txt", delimiter=',', names=['x', 'y'])
depData = np.genfromtxt("../hdi-china-1870-2015.txt", delimiter=',', names=['x', 'y'])
plt.figure(figsize=[10, 5])
plt.title("China : 1870 - 2015")
plt.plot(lifeData['x'], lifeData['y']*0.005, label="Lift Expectancy from Birth")  #这里乘上0.005主要是为了让两个数据尺度保持一致
plt.tick_params(axis='x', rotation=70) #设置x轴标签旋转
plt.plot(depData['x'], depData['y'], label='Human Development Index') #绘制人类发展指数数据
plt.legend() #显示出上面设置的label值
plt.show()
```
